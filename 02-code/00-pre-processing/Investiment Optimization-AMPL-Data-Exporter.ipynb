```python
filters = [
#     {'on' : 'feature', 'min' : 0.0, 'max' : '0.0'}
    {'on' : 'Liquidez', 'min' : 10000},
    {'on' : 'P/VP', 'min' : 0.8}
]
```

```python
def convert_to_float(str_val, default=0.0):
    try:
        str_val = re.sub('[^0-9\.,]','', str_val)
        str_val = str_val.replace('.', '').replace(',', '.')
        return float(str_val)
    except Exception as e:
        print(e, str_val)
        return default
```

```python
def apply_filter(row, filters):
    flag = True
    for f in filters:
        if('min' in f):
            flag = flag and row[f['on']] >= f['min']
    return flag
```

---

```python
columns = [
    'Cotação',
    'FFO Yield',
    'Dividend Yield',
    'P/VP',
    'Valor de Mercado',
    'Liquidez',
    'Preço do m2',
    'Aluguel por m2',
    'Cap Rate',
    'Vacância Média'
]
for col in columns:
    fiis_all[col] = fiis_all[col].apply(lambda v : convert_to_float(v))
#     print('all good with: ', col)
```

## AMPL EXPORT DATA MODULE

```python
filename = 'portifolio.dat'
with open(filename, 'w') as f:
    f.writelines("# PORTIFOLIO ALLOCATION PROBLEM - DATA\n")
    f.writelines(f"param n_assets := {fiis.shape[0]};\n")
    price = "param asset_price :=\n"
    dy = "param asset_dy :=\n"
    liquidity = "param asset_lqdt :=\n"
    pvp = "param asset_pvp :=\n"
    for i, r in fiis.iterrows():
        price     += f"{i+1}\t{r['Cotação']}\n"
        dy        += f"{i+1}\t{r['Dividend Yield']}\n"
        liquidity += f"{i+1}\t{r['Liquidez']}\n"
        pvp       += f"{i+1}\t{r['P/VP']}\n"
    
    f.writelines(f"{price};\n")
    f.writelines(f"{dy};\n")
    f.writelines(f"{liquidity};\n")
    f.writelines(f"{pvp};\n")
    f.writelines(f"\n")
```
