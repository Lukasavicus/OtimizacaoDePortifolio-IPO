```python
def get_data(save=False):
    # --- Do the request -----------------------------------------
    req = requests.get('https://www.fundsexplorer.com.br/ranking')
    soup = bs(req.content, 'html.parser')
    # --- Get the data table -------------------------------------
    table = soup.find('table', {'id' : 'table-ranking'})
    # --- Get headers --------------------------------------------
    headers = [th.text for th in table.find_all('th')]
    # --- Get data -----------------------------------------------
    data = []
    trs = table.find('tbody').find_all('tr')
    for tr in trs:
        tds = tr.find_all('td')
        tds_data = []
        for td in tds[2:]:
            added = False
            try:
                tds_data.append(td['data-order'])
                added = True
            except:
                ...
            if(not added):
                try:
                    tds_data.append(td['data-valor'])
                except:
                    tds_data.append(None)
        d = (
            tds[0]['data-index'],
            tds[1].text,
            *tds_data
        )
        data.append(d)
    fiis_data = pd.DataFrame(columns=headers, data=data)
    if(save):
        fiis_data.to_csv('./bases/funds_explorer_database.csv')
    return fiis_data
```

```python
need_to_get_data = True
fiis_data_loaded = None
if(need_to_get_data):
    fiis_data_loaded = get_data()
```

```python
# Transform Historical Asset Data
def pivot_assets(assets, value_fields):
    date = assets['base_date'].apply(lambda dt : datetime.strptime(dt, '%Y-%m-%d'))
    assets['date'] = date

    year = assets['base_date'].apply(lambda dt : int(dt[0:4]))
    month = assets['base_date'].apply(lambda dt : int(dt[5:7]))

    yearmonth = assets['base_date'].apply(lambda dt : dt[0:4]+dt[5:7])
    assets['yearmonth'] = yearmonth

    # f_dy = assets['dy'] > 0.0

    # value_field = 'yield'
    # value_field = 'dy'
    value_field = value_fields

    f_y = assets[value_field] > 0.0
    f_has_dates = assets['date'] >= '2021-06-01'
    filter_ = f_y & f_has_dates
    fdf = assets[filter_]

    selected_assets = fdf['asset'].unique()#[:100]

    f_asset = fdf['asset'].apply(lambda asset: asset in selected_assets)
    ffdf = fdf[f_asset]

    ffdf_pivoted = ffdf.pivot_table(values = value_field, index = 'yearmonth', columns = 'asset').fillna(0)
    
    return ffdf_pivoted
```

```python
# Pre-processing some variables that would be very difficult to do in AMPL:

fdf = fiis.sort_values(by='dy', ascending=False).reset_index(drop=True)
fdf.rename(columns={'base_quotation':'price'}, inplace=True)
budget = 10000
max_diversity = sum(fdf[fdf['price'] > 0].sort_values(by='price').cumsum()['price'] < budget)
# --------------------------------------------------------------------------------------------------------
dy_price = fdf.apply(lambda row: row['dy']/row['price'] if row['dy'] > 0.0 else 0.0, axis=1)
fdf['dy_price'] = dy_price
top_invest = fdf[['asset', 'price', 'dy', 'dy_price']].sort_values(by='dy_price', ascending=False).reset_index(drop=True)
f = (top_invest['price'] > 0) & (top_invest['dy'] > 0) & (top_invest['price'] < budget)

idx = 0
max_profit = (budget // top_invest[f].loc[idx, 'price']) * top_invest[f].loc[idx, 'dy']

idx = top_invest[f].shape[0] -1
min_profit = (budget // top_invest[f].loc[idx, 'price']) * top_invest[f].loc[idx, 'dy']

#top_invest[f].loc[-1, 'price']
```

```python
filename = '../../01-data/portifolio_data.dat'
sep = '\t'
with open(filename, 'w') as f:
    f.writelines("# PORTIFOLIO ALLOCATION PROBLEM - DATA\n")
    # --- FIIS -----------------------------------------------------------------------------------------------
    f.writelines(f"param n_assets := {fiis.shape[0]};\n")
    f.writelines(f"param asset_data: {sep.join(fiis.columns[1:])} :=\n")
    for i in range(fiis.shape[0]):
        f.writelines(f"{i+1} {sep.join([str(v) for v in fiis.loc[i].values[1:]])} #{fiis.loc[i].values[0]}\n")
    f.writelines(f";\n")
    # --- ASSETS ---------------------------------------------------------------------------------------------
    set_A = "\t".join([str(i+1) for i,_ in enumerate(final_assets.columns)])
    f.writelines(f"set A := {set_A};\n")
    set_T = "'"+"'\t'".join(list(final_assets.index))+"'"
    f.writelines(f"set T := {set_T};\n")
    f.writelines(f"param R: {set_A} :=\n")
    for k, row in final_assets.iterrows():
        values = row.values
        #print(values)
        f.writelines(f"'{k}' {sep.join([str(v) for v in values])}\n")
    f.writelines(f";\n")
    f.writelines(f"param max_diversity = {max_diversity};\n")
    f.writelines(f"param min_profit = {min_profit};\n")
    f.writelines(f"param max_profit = {max_profit};\n")
```
